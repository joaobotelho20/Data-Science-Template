```python


# Caminho para o arquivo
caminho = path + "/GH_Avelar/2003/2003.json"

with open(caminho, 'r', encoding='utf-8') as file:
    data = json.load(file)

# Initialize list to store the extracted data
matches = []

# Check if data is a list or dictionary
if isinstance(data, list):
    # If data is already a list of matches
    partidas = data
elif isinstance(data, dict) and 'partidas' in data:
    # If data is a dictionary with a 'partidas' key
    partidas = data['partidas']
else:
    # If data is in some other format
    print("Unexpected data format")
    partidas = []

# Loop through each match
for partida in partidas:
    # Extract basic match information
    match_info = {
        'mandante': partida['mandante'],
        'visitante': partida['visitante'],
        'data_partida': partida['data_partida'],
        'hora_partida': partida['hora_partida'],
        'placar_mandante': partida['placar_mandante'],
        'placar_visitante': partida['placar_visitante'],
        'resultado': partida['resultado'],
        'estadio': partida['estadio']
    }
    
    # Extract home team statistics (flatten the nested structure)
    for key, value in partida['pontuacao_geral_mandante'].items():
        match_info[f'mandante_{key}'] = value
    
    # Extract away team statistics (flatten the nested structure)
    for key, value in partida['pontuacao_geral_visitante'].items():
        match_info[f'visitante_{key}'] = value
    
    matches.append(match_info)

# Create a DataFrame from the extracted data
df = pd.DataFrame(matches)

# Display the DataFrame
print(df.head())
```
